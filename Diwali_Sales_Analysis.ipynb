## Exploratory Data Analysis

### Gender

```python
df.columns
```

## Count Plot of Customer Gender Distribution:
#### Created a count plot to visualize the distribution of customers by gender and adds labels to each bar to show the count of       observations.

```python
ax= sns.countplot(x= 'Gender', data= df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False)
```

## Total Sales Amount by Gender
#### It shows the total sales amount aggregated by gender, and visualized using a bar plot.

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False)

sns.barplot(x='Gender',y= 'Amount',data= sales_gen) 
```

##### From above groups we can see that most of the buyers are females and even the purchasing power of females are greater than men 

# Age

```python
df.columns
```

### Age Group Distribution of Customers with Gender Breakdown
#### It describes a count plot showing the distribution of customers across different age groups, with a further breakdown by gender.

```python
ax= sns.countplot(data= df, x= 'Age Group', hue= 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

### Total Sales Amount by Age Group
#### It describes the analysis, indicating that the plot will show the total sales amount aggregated by different age groups.

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data= sales_age)
```

###### from above graph we can see that most of the buyers are of age group between 26-35 years female

# State

```python
df.columns
```

## Total Number of Orders from Top 15 States
#### This plot displays the top 15 states in terms of the number of orders received. By analyzing this data, we can identify which states contribute most significantly to the total order volume. This visualization helps in understanding regional demand and market penetration.

```python
# total number of orders from top states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(15)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x= 'State', y='Orders')

plt.xticks(rotation=45)
plt.show()
```

## Total Number of Orders from Top 10 States 
#### This plot shows the total number of orders from the top 10 states, based on the number of orders.

```python
# total amount/sales from top 10 states

sales_state= df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10) 

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x='State', y='Amount')

plt.xticks(rotation=45)
plt.show()
```

###### from above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively.

# Marital Status

## Customer Marital Status Distribution

# Occupation 

## Customer Occupation Distribution Analysis
#### This visualization presents a breakdown of customer occupations, providing insights into the distribution of customers across different occupational categories. Each bar represents the count of customers within a specific occupation, allowing for a visual examination of occupation diversity. The x-axis labels have been rotated for better readability.

```python
sns.set(rc={'figure.figsize':(20,5)})
ax= sns.countplot(data= df, x = 'Occupation') 

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.xticks(rotation=45) 
plt.show()
```

### Total Sales Amount by Occupation
#### This visualization illustrates the total sales amount generated by customers across various occupational categories. Each bar represents the total sales amount attributed to a specific occupation, providing insights into the contribution of different occupations to overall sales. The figure size has been adjusted for optimal presentation, allowing for a clear and informative analysis of sales distribution across occupations.

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x='Occupation', y='Amount')
```

###### From above graph we can see that most of the buyers are working in IT, Aviation and Healthcare sector

# Product Category

## Product Category Distribution Analysis
#### This visualization displays the distribution of products across different categories. Each bar represents the count of products within a specific category, providing insights into the popularity and variety of products available. The x-axis labels have been rotated for better readability. Adjustments to the figure size ensure that the plot is presented clearly and informatively. 

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data= df, x= 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
 
plt.xticks(rotation=45)
plt.show()
```

### Top 10 Product Categories by Total Sales Amount 
#### This visualization illustrates the total sales amount generated by the top 10 product categories. Each bar represents the total sales amount for a specific product category, highlighting the categories that contribute the most to overall sales. The figure size has been adjusted to ensure a clear and detailed presentation of the data. This analysis provides valuable insights into the most profitable product categories, helping to identify key areas of success within the sales data.

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x= 'Product_Category', y='Amount')
```

###### From above graph we can see that most of the sold products are from Food, Footwear, and Electronic category

### Top 10 Most Ordered Products by Product ID
#### This visualization presents the top 10 products based on the number of orders. Each bar represents the total number of orders for a specific product, identified by its Product ID. The figure size has been set to ensure clarity and detail in the presentation. This analysis provides insights into the most popular products, helping to identify which items are in highest demand among customers.

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x= 'Product_ID', y='Orders')
```

### Top 10 Most Sold Products by Product ID
#### This bar chart displays the top 10 products based on the total number of orders. Each bar represents the total orders for a specific product, identified by its Product ID, highlighting the most frequently sold items. The figure size has been set to 12 by 7 inches for a detailed and clear presentation. The x-axis labels have been rotated for better readability, ensuring that each Product ID is easily distinguishable. This analysis offers valuable insights into the highest-selling products, showcasing customer preferences and popular items in the sales data.

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')

plt.xticks(rotation=45)
plt.show()
```
